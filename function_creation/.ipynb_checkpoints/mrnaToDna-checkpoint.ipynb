## Function for Converting mRNA to DNA

```python
def mrnaToDna(mRNAseq):
    """
    converts mRNA sequence to DNA sequence and returns DNA sequence in ATGC format
    input: mRNAseq = mRNA sequence in form of UAGC
    
    output: DNA sequence in the form of ATGC
    """
    mRNAseq=mRNAseq.replace("\n","")
    mapping = mRNAseq.maketrans("UACG","ATGC")
    DNAseq = mRNAseq.translate(mapping)
    
    return DNAseq
```
