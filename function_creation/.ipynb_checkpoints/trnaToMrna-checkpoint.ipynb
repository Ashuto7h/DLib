## Function for Converting tRNA to mRNA

```python
def trnaToMrna(tRNAseq):
    '''
    Converts a tRNAseq to mRNAseq
    
    Input: tRNA sequence in UAGC format
    Output: mRNA sequence in UAGC format
    '''
    tRNAseq=tRNAseq.replace("\n", "")
    #tRNA to mRNA
    mapping_for_mRNA = tRNAseq.maketrans("AUCG","UAGC")
    mRNAseq = tRNAseq.translate(mapping_for_mRNA)
    
    return mRNAseq
```
