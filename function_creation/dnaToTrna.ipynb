## Function for Translation of DNA to tRNA

```python
def dnaToTrna(dnaseq):
    """
    converts DNA sequence to mRNA then to tRNA sequence and returns tRNA sequences
    input: dnaseq = dna sequence in form of ATGC
    
    output: tRna sequence in the form of UACG
    """
    dnaseq=dnaseq.replace("\n","")
    
    mapping_for_mRNA = DNAseq.maketrans("ATGC","UACG")
    mRNAseq = DNAseq.translate(mapping_for_mRNA)
    
    mapping_for_tRNA = mRNAseq.maketrans("UAGC","AUCG")
    tRNAseq = mRNAseq.translate(mapping_for_tRNA)
    
    return tRNAseq
```
