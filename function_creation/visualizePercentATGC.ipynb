## Function for visualizing the ATGC percent in DNA sequence

```python
def visualizePercentATGC(dnaseq,plottype,color='maroon',dpi=100):
    """
    Pass a DNA sequence in ATGC format, will return a Graph ('bar graph', 'pie', 'line graph') based on plot type
    Change color by setting color='colorname/colorcode', by default='maroon'
    Change size of graph by setting dpi=integernumber, by default=100
    
    Input: DNA sequence in ATGC format(required), plottype=['bar','line','pie'](required), color(optional), dpi(optional)
    Output: Matplotlib graph object
    """
    
    dnaseq=dnaseq.replace("\n", "")
    dnaseq=list(dnaseq)

    total_nucleotide=len(dnaseq)
    percentofATGC={}

    for i in dnaseq:
        percentofATGC[i]=dnaseq.count(i)

    for i in percentofATGC:
        percentofATGC[i]=(percentofATGC[i]/total_nucleotide)*100

    atgc=['A','T','G','C']
    atgcpercent=[percentofATGC['A'],percentofATGC['T'],percentofATGC['G'],percentofATGC['C']]
    
    import matplotlib.pyplot as plt
    
    plt.figure(dpi=dpi)
    if(plottype=='bar'):
        plt.bar(atgc, atgcpercent, color = color)
        plt.xlabel("Nitrogenous Bases")
        plt.ylabel("Percentage in DNA")
        plt.title("ATGC Visualization")
        return plt
    elif(plottype=='pie'):
        plt.pie(atgcpercent, labels = atgc, startangle = 90,autopct='%1.0f%%')
        plt.title("ATGC Visualization")
        return plt
    elif(plottype=='line'):
        plt.plot(atgc,atgcpercent,color=color)
        plt.xlabel("Nitrogenous Bases")
        plt.ylabel("Percentage in DNA")
        plt.title("ATGC Visualization")
        return plt
    else:
        return "Available Graphs ['bar','line','pie']"
    
```
