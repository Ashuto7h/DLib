## Function for Converting tRNA to DNA

```python
def trnaToDna(tRNAseq):
    """
    converts tRNA sequence to DNA sequence and returns DNA sequence in ATGC format
    input: tRNAseq = tRNA sequence in form of UAGC
    
    output: DNA sequence in the form of ATGC
    """
    tRNAseq=tRNAseq.replace("\n","")
    
    #tRNA to mRNA
    mapping_for_mRNA = tRNAseq.maketrans("AUCG","UAGC")
    mRNAseq = tRNAseq.translate(mapping_for_mRNA)
    
    #mRNA to DNA
    mapping = mRNAseq.maketrans("UACG","ATGC")
    DNAseq = mRNAseq.translate(mapping)
    
    return DNAseq
```
