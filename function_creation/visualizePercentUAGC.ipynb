## Function for visualizing the content of UAGC in RNA sequence

```python
def visualizePercentUAGC(rnaseq,plottype,color='maroon',dpi=100):
    """
    Pass a RNA sequence in UAGC format, will return a Graph ('bar graph', 'pie', 'line graph') based on plot type
    Change color by setting color='colorname/colorcode', by default='maroon'
    Change size of graph by setting dpi=integernumber, by default=100
    
    Input: RNA sequence in UAGC format(required), plottype=['bar','line','pie'](required), color(optional), dpi(optional)
    Output: Matplotlib graph object
    """
    
    rnaseq=rnaseq.replace("\n", "")
    rnaseq=list(rnaseq)

    total_nucleotide=len(rnaseq)
    percentofUAGC={}

    for i in rnaseq:
        percentofUAGC[i]=rnaseq.count(i)

    for i in percentofUAGC:
        percentofUAGC[i]=(percentofUAGC[i]/total_nucleotide)*100

    uagc=['U','A','G','C']
    uagcpercent=[percentofUAGC['U'],percentofUAGC['A'],percentofUAGC['G'],percentofUAGC['C']]
    
    import matplotlib.pyplot as plt
    
    plt.figure(dpi=dpi)
    if(plottype=='bar'):
        plt.bar(uagc, uagcpercent, color = color)
        plt.xlabel("Nitrogenous Bases")
        plt.ylabel("Percentage in RNA")
        plt.title("UAGC Visualization")
        return plt
    elif(plottype=='pie'):
        plt.pie(uagcpercent, labels = uagc, startangle = 90,autopct='%1.0f%%')
        plt.title("UAGC Visualization")
        return plt
    elif(plottype=='line'):
        plt.plot(uagc,uagcpercent,color=color)
        plt.xlabel("Nitrogenous Bases")
        plt.ylabel("Percentage in RNA")
        plt.title("UAGC Visualization")
        return plt
    else:
        return "Available Graphs ['bar','line','pie']"
```
