```python
# Given 'n', the no. of lobes, we need to print the double-helix structure of DNA
# Lobes - lobes in a nucleus are made of chromatin, a mixture of DNA and proteins.

# DNA primarily consists of 4 hydrocarbons i.e. cytosine [C], guanine [G], adenine[A], thymine [T].
# DNA bases pair up with each other, A with T and C with G, to form units called base pairs.



# imported math module
import math


# Storing combinations in a list DNA
DNA = [ "ATTAATTA", "TAGCTAGC", "CGCGATAT", "TAATATGC", "ATCGTACG", "CGTAGCAT" ]


# function defined to print upper half of the DNA or the upper lobe
def print_upper_half(str):
    first=0
    second=0
    pos = 0
     
    # Each half of the DNA is made of combination of two compounds
    for i in range(1,5):
        # Taking the two carbon compounds from the string
        first = str[pos]
        second = str[pos+1]
        pos += 2
          
        for  j in range ( 4 - i, 0,-1):
            print(" ",end="")
        print(first,end="")
        for  j in range (1, i):
            print("--",end="")
        print(second)
        

# function defined to print lower half of the DNA or the lower lobe
def print_lower_half(str):
    first=0
    second=0
    pos = 0
    for i in range(1,5):
        first = str[pos]
        second = str[pos+1]
        pos += 2
          
        for  j in range(1,i):
            print(" ",end="")
        print(first,end="")
        for  j in range (4 - i, 0,-1):
            print("--",end="")
        print(second)
        
        
# main function defined to print 'n' parts of DNA
def print_double_helix_DNA( str,  n):
    for i in range(0,n):
        x = i % 6
        # Calling print_upper_half and printing the upper half of double-helix
        if (x % 2 == 0):
            print_upper_half(str[x])
        else:
            # Calling print_lower_half and printing the lower half of double-helix
            print_lower_half(str[x])
            

            
#TESTING
# Assuming n
n = 10

# calling the print function
print_double_helix_DNA( DNA,  n)
```
