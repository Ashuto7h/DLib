## Function for Visualizing Protein with RNA sequence

```python
def visualizeProteinWithRna(rnaseq,plottype,color='maroon',dpi=100):
    """
    Pass a RNA sequence in UACG format and then it will convert it into a Protein sequence out of RNA sequence
    Change color by setting color='colorcode/name', By default = 'maroon'
    Change size of graph by setting dpi= integer_value, By default=100
    
    Input: RNA sequence in UACG format(required), plottype(required), color(optional), dpi(optional)
    Output: Matplotlib Graph object
    """
    import matplotlib.pyplot as plt
    
# conversion of RNA to Protein sequence
    rnaseq=rnaseq.replace("\n", "")
    RNACodons = {
        "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
        "UGU": "C", "UGC": "C",
        "GAU": "D", "GAC": "D",
        "GAA": "E", "GAG": "E",
        "UUU": "F", "UUC": "F",
        "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G",
        "CAU": "H", "CAC": "H",
        "AUA": "I", "AUU": "I", "AUC": "I",
        "AAA": "K", "AAG": "K",
        "UUA": "L", "UUG": "L", "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
        "AUG": "M",
        "AAU": "N", "AAC": "N",
        "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
        "CAA": "Q", "CAG": "Q",
        "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
        "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S", "AGU": "S", "AGC": "S",
        "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
        "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
        "UGG": "W",
        "UAU": "Y", "UAC": "Y",
        "UAA": "*", "UAG": "*", "UGA": "*"
    }
    
    protein = ""
    for i in range(0, len(rnaseq), 3):
        codon = rnaseq[i:i+3]
        protein += RNACodons[codon]
    
    proteins=list(RNACodons.values())
    proteincount={}
    for i in proteins:
        proteincount[i]=proteins.count(i)
    
# conversion of RNA to Protein sequence

    aminoacidname=list(proteincount.keys())
    aminoacidcount=list(proteincount.values())
    
    
    plt.figure(dpi=dpi)
    
    if(plottype=='bar'):
        plt.bar(aminoacidname, aminoacidcount, color =color)
        plt.xlabel("Amino Acid")
        plt.ylabel("Amino Acid Count")
        plt.title("Protein Visualization")
        return plt
    elif(plottype=='pie'):
        plt.pie(aminoacidcount, labels = aminoacidname, startangle = 180)
        plt.title("Protein Visualization")
        return plt
    elif(plottype=='line'):
        plt.plot(aminoacidname, aminoacidcount,color=color)
        plt.xlabel("Amino Acid")
        plt.ylabel("Amino Acid Count")
        plt.title("Protein Visualization")
        return plt
    else:
        return "Available Graphs ['bar','line','pie']"
    
```
