## Function for Converting mRNA to tRNA

```python
def mrnaToTrna(mRNAseq):
    '''
    Converts a mRNAseq to tRNAseq
    
    Input: mRNA sequence in UAGC format
    Output: tRNA sequence in UAGC format
    '''
    mRNAseq=mRNAseq.replace("\n", "")
    #from mRNA to tRNA
    mapping_for_tRNA = mRNAseq.maketrans("UAGC","AUCG")
    #to tRNA
    tRNAseq = mRNAseq.translate(mapping_for_tRNA)
    
    return tRNAseq
```
